```python
base_dir = '/content/drive/MyDrive/Colab Notebooks/Udacity/Spill Detection - CNN/Data'
train_dir = os.path.join(base_dir,'train')
test_dir = os.path.join(base_dir,'test')
```

```python
IMG_Size = 64
BATCH_SIZE = 32
EPOCHS = 30
```

```python
from keras import Input, layers, models, regularizers

input_layer = Input(shape = (IMG_Size,IMG_Size,3))
x = layers.Conv2D(128,(3,3),padding='same')(input_layer)
x = layers.Activation('relu')(x)
x = layers.Conv2D(128,(3,3),padding='same')(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(128,(3,3),padding = 'same')(x)
x = layers.Activation('relu')(x)

x = layers.Conv2D(256,(3,3),padding='same')(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(256,(3,3),padding='same')(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(256,(3,3),padding = 'same')(x)
x = layers.Activation('relu')(x)

x = layers.Flatten()(x)
x = layers.Dense(512,activation='relu')(x)

output_layer = layers.Dense(1,activation='sigmoid')(x)
model = models.Model(input_layer,output_layer)
model.summary()
```

```python
from keras import optimizers

model.compile(loss = 'binary_crossentropy',
              optimizer = optimizers.Adam(),
              metrics=['acc'])
```

**Data Augmentation**

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=5,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    validation_split=0.2,
    horizontal_flip=True)

train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_Size, IMG_Size),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_Size, IMG_Size),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
train_generator.class_indices
```

```python
hist = model.fit(train_generator,
                 epochs = EPOCHS,
                 validation_data = validation_generator)
```

```python
acc = hist.history["acc"]
val_acc = hist.history["val_acc"]
loss = hist.history["loss"]
val_loss = hist.history["val_loss"]

epochs = range(1, len(acc) + 1)

# accuracy
fig = plt.figure(figsize=(25,8))
ax1 = fig.add_subplot(121)
ax1.plot(epochs, acc, "bo", label="Training acc")
ax1.plot(epochs, val_acc, "b-", label="Validation acc")
ax1.set_title("Training and validation accuracy")
ax1.legend()

# loss
ax2 = fig.add_subplot(122)
ax2.plot(epochs, loss, "ro", label="Training loss")
ax2.plot(epochs, val_loss, "r-", label="Validation loss")
ax2.set_title("Training and validation loss")
ax2.legend()

fig.savefig('/content/drive/MyDrive/Colab Notebooks/Udacity/Spill Detection - CNN/Data/fig.png')
```
